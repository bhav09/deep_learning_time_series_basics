```
import pandas as pd
import numpy as np
```

```
data=pd.read_csv('https://raw.githubusercontent.com/jbrownlee/Datasets/master/daily-total-female-births.csv')
```

```
data.head()
```

```
len(data.index)
```

```
x=data['Date']
y=data['Births']
```

```
x
```

```
y
```

```
#converting the x var to unix time 
import datetime
import time
conversion=x
for i in range(len(x)):
  l=conversion[i].split('-')
  #print(type(l[0]))
  t=datetime.date(int(l[0]),int(l[1]),int(l[2]))
  unixtime=time.mktime(t.timetuple())
  conversion[i]=unixtime
x=conversion
print(x)
```

```
#this is an example of univariate multilayer perceptron 
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```
model=Sequential()
```

```
model.add(Dense(units=10,activation='relu',input_dim=1))
model.add(Dense(units=1))
```

```
model.compile(optimizer='adam',loss='mse')
```

```
x=np.asarray(x)
print(x.shape)
x=np.expand_dims(x,-1)
print(x.shape)
```

```
x=x.astype(np.float32)
model.fit(x,y,epochs=500,verbose=1)
```

```
x_input='1960-01-01'
k=x_input.split('-')
print(x_input,'\t',k)
t=datetime.date(int(k[0]),int(k[1]),int(k[2]))
unixtime1=time.mktime(t.timetuple())
print(type(unixtime1))
x_input=unixtime1
x_input
x_input=np.array(x_input,dtype=float)
```
